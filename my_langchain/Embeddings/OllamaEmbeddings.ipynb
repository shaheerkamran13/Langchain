# Ollama Embeddings - Open Source Embeddings

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings = (
    OllamaEmbeddings(model="gemma2:2b") # by default it used llama2
)
```

```python
embeddings
```

```python
r1 = embeddings.embed_documents([
    "Alpha is the first letter of greek alphabet",
    "Beta is the second letter of greek alphabet",
])
```

```python
r1
```

```python
r2 = embeddings.embed_query("Beta is the second letter of greek alphabet")
```

```python
r2
```

```python
# Other embedding Models

# https://ollama.com/blog/embedding-models

embeddings = OllamaEmbeddings(model="mxbai-embed-large") # Download this opensource model
text = 'This is a test document'
query_result =  embeddings.embed_query(text)

query_result
```
