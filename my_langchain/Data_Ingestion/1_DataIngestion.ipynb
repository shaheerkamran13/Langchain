```python
# Reading a PDF file

from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader('sample.pdf')

loader

docs = loader.load()

docs

# type(docs[0])
```

```python
# Web Based Loader
import bs4
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader(web_paths=("https://www.freecodecamp.org/news/how-to-optimize-for-change-software-development/",),
                        bs_kwargs=dict(parse_only = bs4.SoupStrainer(
                            class_ = ("trending-guides")
                        ))
                    )
```

```python
loader.load()
```

```python
# Arxiv Loader

from langchain_community.document_loaders import arxiv, ArxivLoader

loader = ArxivLoader(
    
    query = "2603.01607",
    load_max_docs = 2
)


```

```python
loader.load()
```
