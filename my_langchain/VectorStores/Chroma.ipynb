# Chroma DB

Chroma is a AI-native open source vector database focused on developer productivity and happiness. Chroma DB is licensed under Apache 2.0

```python
# building a sample vectordb

from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter

```

```python
loader = TextLoader("speech.txt")

data = loader.load()

data
```

```python
# Text Splitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap = 0)

splits =  text_splitter.split_documents(data)

splits
```

```python
# embeddings 

embeddings = OllamaEmbeddings(model="mxbai-embed-large:latest")
```

```python
vectorDB = Chroma.from_documents(documents = splits, embedding = embeddings)

vectorDB
```

```python
# Query it

query = "What is the Speech is study of?"

docs = vectorDB.similarity_search(query)

docs[0].page_content
```

```python
## Save to the disk

vectorDB = Chroma.from_documents(documents = splits, embedding = embeddings, persist_directory = "./chroma_db")

```

```python
# load from disk

db2 = Chroma(persist_directory="./chroma_db", embedding_function=embeddings)

docs = db2.similarity_search_with_score(query=query)
```

```python
docs
```

```python
# Retriever option

retriever = vectorDB.as_retriever()
retriever.invoke(query)[0].page_content
```
