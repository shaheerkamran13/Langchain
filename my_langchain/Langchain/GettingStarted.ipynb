```python
import os
from dotenv import load_dotenv

load_dotenv()

os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")

# Langsmith Tracking
os.environ["LANGCHAIN_API_KEY"] = os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_TRACING_V2"] = "true"

os.environ["LANGCHAIN_PROJECT"] = os.getenv("LANGCHAIN_PROJECT")
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model="gpt-4o"
)

print(llm)
```

```python
# Input and get response from LLM
result = llm.invoke("What is Generative AI")

result
```

```python
### ChatPrompt Template


from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    [
        ("system","You are an expert AI Engineer. Provide me some answers based on the question"),
        ("user", "{input}")
    ]
)


prompt
```

```python
chain = prompt|llm

response=chain.invoke({"input":"Can you tell me something about Langsmith?"})

print(response)
```

```python
type(response)
```

```python
# Output pasrser: To display the response
# The default outputparser being used would be:- stroutput Parser

# stroutput Parser ::  Significance: Getting the message from LLM and how to display it



from langchain_core.output_parsers import StrOutputParser
output_parser = StrOutputParser()
chain = prompt| llm | output_parser

response=chain.invoke({"input":"Can you tell me something about Langsmith?"})
print(response)
```
